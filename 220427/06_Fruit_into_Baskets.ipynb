```python
class Solution:
    def totalFruit(self, fruits):

        # 예외처리: 두 종류 이하이면 전부 다 챙길 수 있음
        set_fruits = set(fruits)
        if len(set_fruits) <= 2:
            return len(fruits)
        
        # 답 후보들 저장 list
        cands = []

        # start point 설정
        for i in range(len(fruits)):
            types = []                  # 몇 종류인지 체크하기
            types.append(fruits[i])     # start point 넣고 시작
            mine = 1                    # 챙긴 개수

            # 뒤로 쭉 가면서
            for j in range(i+1, len(fruits)):
                
                # 이미 있는 type이면 넣으면 됨 (mine += 1)
                if fruits[j] in types:
                    mine += 1
                
                # 없는 type이면, type이 아직 1개일 땐 넣고 (mine += 1) type 추가 (types.append)
                elif (fruits[j] not in types) & len(types) == 1:
                    mine += 1
                    types.append(fruits[j])
                    
                # 없는 type인데 type도 이미 두 개면 재시작 (break)
                else:
                    break

            
            # 현재까지 얻은 개수를 (mine) 후보로 넣기 (cands) 
            cands.append(mine)
                
        # 답은 후보들 중 최댓값        
        return max(cands)
```

```python
class Solution:
    def totalFruit(self, fruits):

        # 예외처리: 두 종류 이하이면 전부 다 챙길 수 있음
        set_fruits = set(fruits)
        if len(set_fruits) <= 2:
            return len(fruits)
        
        # max를 위한 답 초기화
        ans = 0
        
        # start point 설정
        for i in range(len(fruits)):
            types = []              # 몇 종류인지 체크하기
            types.append(fruits[i]) # start point 넣고 시작
            basket = 1              # 챙긴 개수

            # 뒤로 쭉 가면서
            for j in range(i+1, len(fruits)):

                # 이미 있는 type이면 넣으면 됨 (basket += 1)
                if fruits[j] in types:
                    basket += 1
                
                # 없는 type이면, type이 아직 1개일 땐 넣고 (basket += 1) type 추가 (types.append)
                elif (fruits[j] not in types) & (len(types) == 1):
                    types.append(fruits[j])
                    basket += 1
                    
                # 없는 type인데 type도 이미 두 개면 재시작 (break)
                else:
                    break

            # start point 뒤로 다 훑었건 break로 나왔건 정답 update
            ans = max(ans, basket)

        return ans
```

# Two Pointers - sliding window

```python
# 모범답안
# ex) [0,1,2,2,3,2]
from collections import defaultdict

class Solution1:
    def totalFruit(self, fruits):
        basket = defaultdict(int)   # 종류 별로 넣을 바구니
        
        ans = 0                     # max를 위한 답 초기화
        left = 0                    # left pointer
        
        # 오른쪽 포인터를 한 칸 씩 이동
        for right, fruit in enumerate(fruits):          # range 쓰면 더 오래 걸림 (len / indexing)

            # 일단 하나 넣고
            basket[fruit] += 1

            # 넣다가 종류가 2개보다 많아지면 왼쪽 포인터를 끌고 옴
            while (len(basket) > 2) & (left < right):
                
                # 왼쪽 포인터가 가리키는 과일을 하나 씩 제거
                basket[fruits[left]] -= 1

                # 해당 종류의 과일이 없어지면 key 제거
                if basket[fruits[left]] == 0:
                    basket.pop(fruits[left], None)
                
                # 왼쪽 포인터 이동
                left += 1
            
            # 정답 update
            ans = max(ans, sum(basket.values()))

        return ans

# 1119ms (41.84%) / 20.1MB (58.84%)
# 1566ms (10.71%) / 20.3MB (39.32%)
# 1431ms (16.04%) / 20.2MB (58.84%)
# avg
# 1372ms / 52.33%
```
